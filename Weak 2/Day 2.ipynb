```python
# Find second minimum value

def second_min(s):
    sorted_set = sorted(s)
    return sorted_set[1]
numbers = {101,102,103,101,102}
print("Second Minimum:", second_min(numbers))
```

```python
# Revert a list
a=[8,77,99,33,22,44]
reversed_list = []

for i in range(len(a)-1, -1, -1):
    reversed_list.append(a[i])

print(reversed_list)
```

```python
#sum & Average
a = [8, 77, 99, 33, 22, 44]
total = 0
for num in a:
    total += num
print("Sum:", total)
average=total/len(a)
print("Average: ",average)
```
